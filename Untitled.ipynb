```python
import pandas as pd
import calendar

cycle_start_day = 'Monday'
cycle_period = 7
```

```python
def find_first_monday(year, month):
    # Get the calendar for the given month and year
    cal = calendar.monthcalendar(year, month)

    # Find the first week that has a Monday
    for week in cal:
        if week[calendar.MONDAY] != 0:
            # Return the date of the first Monday
            return week[calendar.MONDAY]
```

```python
def get_data():
    df = pd.read_excel('/Users/jigme/Desktop/data/sales.xlsx',dtype={'Scan Code':str})
    df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['WeekDay'] = df['Date'].dt.day_of_week
    print('Year(s) : ',df['Year'].unique())
    print('Months(s) : ',df['Month'].unique())
    print('start date: ',df['Date'][0])
    print('end date: ',df['Date'][len(df)-1])
    return df
```

```python
def calculate_cycle_average(year,month,cycle_length,num_cycles,item):
    weekly_sales = []
    start_day = find_first_monday(year,month)
    print('First Monday of the Month: ',start_day)
    _, last_day = calendar.monthrange(year, month)
    for i in range(0,num_cycles):
        week_total = 0
        for j in range(start_day,start_day + cycle_length):

            try:
                week_total = week_total + daily.loc[(month,j,item),'Qty'].values[0]

            except KeyError:
                continue



        start_day = start_day + cycle_length
        if start_day>last_day:
            month = month+1
            initial_start = start_day - cycle_length
            days_accounted = last_day - initial_start
            remaining_days = cycle_length - days_accounted -1
            for k in range(1,remaining_days+1):
                try:
                    week_total = week_total + daily.loc[(month,k,item),'Qty'].values[0]
                except KeyError:
                    continue
                    
        print(f"period: {month}/{start_day} week: {i} total_sales: {week_total}")
        weekly_sales.append(week_total)
    return weekly_sales
```
